### 4

Какова доля выживших после крушения пассажиров? Какова доля мужчин и женщин среди выживших?

```python
survived_quantity = data[data['Survived'] == 1].shape[0]
survived_female_quantity = data[(data['Survived'] == 1) & (data['Sex'] == "female")].shape[0]
survived_male_quantity = data[(data['Survived'] == 1) & (data['Sex'] == "male")].shape[0]
print(data['Survived'].mean(), survived_female_quantity/ survived_quantity, survived_male_quantity/ survived_quantity)

```

### 5

Сколько пассажиров ехало в каждом классе? Кого было больше в самом многолюдном классе — мужчин или женщин?

```python
data_by_class = data.groupby('Pclass')["Pclass"]
print(data_by_class.describe()['count'])

data_by_class_female = data[(data['Pclass'] == 3) & (data['Sex'] == "female")].shape[0]
data_by_class_male = data[(data['Pclass'] == 3) & (data['Sex'] == "male")].shape[0]
if data_by_class_female >= data_by_class_male:
  print("В многолюдном классе больше женщин")
else:
  print("В многолюдном классе больше мужчин")
```

```python
min_fare = data["Fare"].min()
max_fare = data["Fare"].max()
mean_fare = data["Fare"].mean()
print(min_fare, max_fare, mean_fare)

data_S_min = data[data["Embarked"] == "S"]["Fare"].min()
data_S_max = data[data["Embarked"] == "S"]["Fare"].max()
data_S_mean = data[data["Embarked"] == "S"]["Fare"].mean()
print(data_S_min, data_S_max, data_S_mean)

plt.style.use('seaborn-whitegrid')

plt.figure(figsize=(28,15))
bins = np.linspace(0, data["Fare"].max(), 50)
plt.hist(data[data["Embarked"] == "C"]["Fare"], bins=bins, alpha= 0.5, color = '#E69F00')
plt.hist(data[data["Embarked"] == "Q"]["Fare"], bins=bins, alpha= 0.5, color = '#56B4E9')
plt.hist(data[data["Embarked"] == "S"]["Fare"], bins=bins, alpha= 0.5, color = '#F0E442')
plt.title('Распределения стоимостей билетов в зависимости от направления')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.legend(['Cherbourg', 'Queenstown', 'Southampton'])
```
